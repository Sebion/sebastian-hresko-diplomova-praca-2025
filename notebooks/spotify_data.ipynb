```python
import pandas as pd
from IPython.display import display
import json
import requests
import os
import base64
from tqdm import tqdm
from spotify_api import get_token, get_track
from helper import key_to_camelot
import time

# Load the CSV file into a DataFrame
df = pd.read_csv('filtered_music.csv')

# Function to extract the first artist ID from track data
def extract_first_artist_id(track_data):
    first_artist_id = track_data['album']['artists'][0]['id'] if track_data['album']['artists'] else None
    return first_artist_id

# Function to get track data in batches
def get_tracks_in_batches(track_ids, token):
    url = "https://api.spotify.com/v1/tracks"
    headers = {
        "Authorization": f"Bearer {token}"
    }
    params = {
        "ids": ",".join(track_ids)
    }
    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()
    return response.json()['tracks']

# Add a new column to the DataFrame for the first artist ID
token = get_token()
artist_ids = []

# Process track IDs in batches of 50
batch_size = 50
for i in tqdm(range(0, len(df['spotify_id']), batch_size), desc="Processing tracks"):
    batch_ids = df['spotify_id'][i:i + batch_size].tolist()
    track_data_batch = get_tracks_in_batches(batch_ids, token)
    for track_data in track_data_batch:
        artist_id = extract_first_artist_id(track_data)
        artist_ids.append(artist_id)

    # Pause for 30 seconds after every 30 batches
    if (i // batch_size + 1) % 30 == 0:
        time.sleep(30)

df['artist_id'] = artist_ids

# Save the updated DataFrame back to the CSV file
# df.to_csv('filtered_music.csv', index=False)

# Display the updated DataFrame
display(df)
```

```python
import pandas as pd
from IPython.display import display
import requests
from tqdm import tqdm
import time
from spotify_api import get_token

# Load the CSV file into a DataFrame
df = pd.read_csv('filtered_music.csv')

# Function to get artist data in batches
def get_artists_data(artist_ids, token):
    url = "https://api.spotify.com/v1/artists"
    headers = {
        "Authorization": f"Bearer {token}"
    }
    params = {
        "ids": ",".join(artist_ids)
    }
    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()
    return response.json()['artists']

# Add a new column to the DataFrame for genres
token = get_token()
genres = []

# Process artist IDs in batches of 50
batch_size = 50
artist_ids = df['artist_id'].tolist()
for i in tqdm(range(0, len(artist_ids), batch_size), desc="Processing artists"):
    batch_ids = artist_ids[i:i + batch_size]
    artist_data_batch = get_artists_data(batch_ids, token)
    for artist_data in artist_data_batch:
        genres.append(artist_data.get('genres', []))

    # Pause for 30 seconds after every 30 batches
    if (i // batch_size + 1) % 30 == 0:
        time.sleep(30)

df['genres'] = genres

# Save the updated DataFrame back to the CSV file
df.to_csv('filtered_music.csv', index=False)

# Display the updated DataFrame
display(df)
```

```python
import pandas as pd
from mp3_downloader import MP3Downloader  # Import the MP3Downloader class
from tqdm import tqdm
import re

# Function to sanitize file names
def sanitize_filename(filename):
    return re.sub(r'[\\/*?:"<>|]', "", filename)

# Initialize the MP3Downloader
downloader = MP3Downloader(base_directory="../mp3songs")

# Load the CSV file into a DataFrame
df = pd.read_csv('pridaj_edm_a_finito.csv')

# Ensure the 'genres' column is a list
df['genres'] = df['genres'].apply(eval)

# Initialize an empty list to store file paths
file_paths = []

# Iterate over the DataFrame rows with progress tracking
for index, row in tqdm(df.iterrows(), total=df.shape[0], desc="Downloading MP3s"):
    try:
        sanitized_name = sanitize_filename(row['name'])
        file_path = downloader.download_mp3(row['preview'], sanitized_name, playlist_name="48_genres_11_main_genres")
        file_paths.append(file_path)
    except FileNotFoundError as e:
        print(f"Error downloading {row['name']}: {e}")
        file_paths.append(None)  # Append None or a placeholder if the download fails

# Add the file paths to the DataFrame
df['file_path'] = file_paths

# Save the updated DataFrame back to the CSV file
df.to_csv('48_genres_11_main_genres_with_paths.csv', index=False)

# Display the updated DataFrame
display(df)
```
